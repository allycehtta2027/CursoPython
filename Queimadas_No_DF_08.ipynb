Explicação das colunas:

    Data: Data da ocorrência das queimadas (formato Ano-Mes-Dia).
    Estado: Estado onde a queimada ocorreu (DF para Distrito Federal).
    Municipio: Nome do município (Brasília).
    Latitude: Coordenada de latitude do local da queimada.
    Longitude: Coordenada de longitude do local da queimada.
    Focos: Número de focos de queimadas registrados.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Exemplo de leitura de dados (supondo que você já tenha um arquivo CSV com dados de queimadas)
df = pd.read_csv('queimadas em brasilia .csv')

# Exibir as primeiras linhas do DataFrame
print(df.head())

# Exemplo de DataFrame ('queimadas_brasília.csv')
# |Data        |Estado |Municipio    |Latitude      |Longitude           |Focos |
# |------------|-------|-------------|--------------|--------------------|------|
# |2024-01-10, |DF,    |Brasília,     |-15.7801,     |-47.9292,          |4
# |2024-02-05, |DF,    |Brasília,     |-15.7801,     |-47.9292,          |2
# |2024-03-15, |DF,    |Brasília,     |-15.7801,     |-47.9292,          |6
# |2024-04-20, |DF,    |Brasília,     |-15.7801,     |-47.9292,          |1
# |2024-05-25, |DF,    |Brasília,     |-15.7801,     |-47.9292,          |3
# |2024-06-30, |DF,    |Brasília,     |-15.7801,     |47.9292,           |8
# |2024-07-18, |DF,    |Brasília,     |-15.7801,     |-47.9292,          |5
# |2024-08-12, |DF,    |Brasília,     |-15.7801,     |-47.9292,          |10
# |2024-09-22, |DF,    |Brasília,     |-15.7801,     |-47.9292,          |7
# |2024-10-05, |DF,    |Brasília,     |-15.7801,     |-47.9292,          |9
# |2024-11-15, |DF,    |Brasília,     |-15.7801,     |-47.9292,          |2
# |2024-12-28, |DF,    |Brasília,     |-15.7801,     |-47.9292,          |6


# Gráfico de Barras: Focos de Queimadas por Estado
Neste gráfico de barras, estamos usando a função sns.
Barplot para criar um gráfico de barras com base nos dados do DataFrame df.
Bartplot da biblioteca searborn para agrupar os dados por estado e calcular a soma dos focos.

plt.figure(figsize=(10, 6))
sns.barplot(x='Estado', y='Focos', data=df, ci=None, estimator=sum)
plt.xticks(rotation=45)
plt.title('Focos de Queimadas por Estado')
plt.show()

# Gráfico de Linha: Evolução dos Focos ao Longo do Tempo
Neste gráfico, você visualiza a evolução dos focos ao longo do tempo,
com o eixo x representando as datas e o eixo y representando o número de focos.
Ele mostra como os focos de queimadas se comportaram ao longo dos meses em 2024.

df['Data'] = pd.to_datetime(df['Data'])
df_grouped = df.groupby('Data').sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(df_grouped['Data'], df_grouped['Focos'])
plt.title('Evolução dos Focos de Queimadas ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Número de Focos')
plt.grid(True)
plt.show()

# Mapa de Calor: Densidade das Queimadas por Latitude e Longitude
Esse é um mapa de calor que mostra a densidade das queimadas em uma determinada região.
O gráfico usa as colunas de Latitude e Longitude para determinar onde as queimadas ocorreram.
o mapa de calor pode ter uma única região destacada, indicando a maior concentração de queimadas.


plt.figure(figsize=(10, 6))
sns.kdeplot(df['Longitude'], df['Latitude'], cmap="Reds", shade=True, bw_adjust=0.5)
plt.title('Densidade de Queimadas no Brasil')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o DataFrame
df = pd.read_csv('datasets/queimadas_df.csv')

# Exibir as primeiras linhas do DataFrame
print(df.head())

# Gráfico de linha do total de áreas queimadas ao longo do tempo
plt.figure(figsize=(10, 6))
df['data'] = pd.to_datetime(df['data'])  # Certifique-se de que a coluna 'data' está no formato datetime
df.groupby('data')['area_queimada'].sum().plot()
plt.title('Total de Áreas Queimadas ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Área Queimada (hectares)')
plt.grid()
plt.show()

# Gráfico de barras das áreas queimadas por estado
plt.figure(figsize=(10, 6))
df.groupby('estado')['area_queimada'].sum().plot(kind='bar')
plt.title('Áreas Queimadas por Estado')
plt.xlabel('Estado')
plt.ylabel('Área Queimada (hectares)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
