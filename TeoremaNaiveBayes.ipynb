O teorema de Bayes recebe este nome devido ao pastor e matemtico ingls Thomas Bayes (1701  1761), que foi o primeiro a fornecer uma equao que permitiria que novas evidncias atualizassem a probabilidade de um evento a partir do conhecimento a priori (ou a crena inicial na ocorrncia de um evento). O teorema de Bayes foi mais tarde desenvolvido por Pierre-Simon Laplace, que foi o primeiro a publicar uma formulao moderna em 1812 em seu livro Teoria Analtica de Probabilidade, na traduo do francs. Harold Jeffreys colocou o algoritmo de Bayes e a formulao de Laplace em uma base axiomtica. Jeffreys escreveu que "o teorema de Bayes  para a teoria da probabilidade o que o teorema de Pitgoras  para a geometria".

O teorema de Bayes recebe este nome devido ao pastor e matemtico ingls Thomas Bayes (1701  1761), que estudou como calcular a distribuio para o parmetro de probabilidade de uma distribuio binomial (terminologia moderna).

O manuscrito no publicado de Bayes foi editado significativamente por Richard Price antes de ser lido postumamente na Royal Society. Price editou o principal trabalho de Bayes An Essay towards Solving a Problem in the Doctrine of Chances (1763), que aparece em Philosophical Transactions e contm o teorema de Bayes.

Price escreveu uma introduo para o artigo, que fornece algumas das bases filosficas da estatstica bayesiana. Em 1765, Price foi eleito membro da Royal Society em reconhecimento ao seu trabalho sobre o legado de Bayes.
