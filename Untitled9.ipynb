```python
import numpy as np

# Classe para representar um ponto (coordenadas x, y)
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

# Funo para calcular a distncia mdia entre a boca esquerda e direita (Mouth Aspect Ratio - MAR)
def calculo_ear(face, p_boca_esq, p_boca_dir):
    try:
        # Converter a lista de objetos Point para uma lista de coordenadas numpy
        face_coords = np.array([[coord.x, coord.y] for coord in face])

        # Extrair os pontos da boca  esquerda e  direita usando os ndices
        boca_esq_coords = face_coords[p_boca_esq, :]  # Pontos  esquerda da boca
        boca_dir_coords = face_coords[p_boca_dir, :]  # Pontos  direita da boca

        # Calcular a mdia das coordenadas dos pontos  esquerda e  direita
        boca_esq_mean = boca_esq_coords.mean(axis=0)  # Mdia das coordenadas  esquerda
        boca_dir_mean = boca_dir_coords.mean(axis=0)  # Mdia das coordenadas  direita

        # Calcular a distncia entre as mdias dos pontos da boca
        dist_esq_dir = np.linalg.norm(boca_esq_mean - boca_dir_mean)  # Distncia euclidiana

        # Retornar a distncia, que pode ser interpretada como o MAR
        return dist_esq_dir

    except Exception as e:
        print(f"Ocorreu um erro: {e}")
        return None

# Exemplo de uso da funo com dados fictcios
# Definindo alguns pontos fictcios
face = [
    Point(100, 100)
    Point(140, 100)
    Point(180, 100)
    # Pontos da boca (exemplo de pontos fictcios  esquerda e  direita)
    Point(385, 380), Point(387, 373)
]

# ndices de pontos da boca
p_boca_esq = [0, 1, 2, 3]  # ndices fictcios para a boca esquerda
p_boca_dir = [8, 9, 10, 11]  # ndices fictcios para a boca direita

# Chamar a funo para calcular o MAR
distancia = calculo_ear(face, p_boca_esq, p_boca_dir)

print(f"A distncia (MAR) entre a boca esquerda e direita : {distancia}")
```
