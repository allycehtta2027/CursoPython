```python
import numpy as np

# Classe para representar um ponto (coordenadas x, y)
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

# Função para calcular a distância média entre a boca esquerda e direita (Mouth Aspect Ratio - MAR)
def calculo_ear(face, p_boca_esq, p_boca_dir):
    try:
        # Converter a lista de objetos Point para uma lista de coordenadas numpy
        face_coords = np.array([[coord.x, coord.y] for coord in face])

        # Extrair os pontos da boca à esquerda e à direita usando os índices
        boca_esq_coords = face_coords[p_boca_esq, :]  # Pontos à esquerda da boca
        boca_dir_coords = face_coords[p_boca_dir, :]  # Pontos à direita da boca

        # Calcular a média das coordenadas dos pontos à esquerda e à direita
        boca_esq_mean = boca_esq_coords.mean(axis=0)  # Média das coordenadas à esquerda
        boca_dir_mean = boca_dir_coords.mean(axis=0)  # Média das coordenadas à direita

        # Calcular a distância entre as médias dos pontos da boca
        dist_esq_dir = np.linalg.norm(boca_esq_mean - boca_dir_mean)  # Distância euclidiana

        # Retornar a distância, que pode ser interpretada como o MAR
        return dist_esq_dir

    except Exception as e:
        print(f"Ocorreu um erro: {e}")
        return None

# Exemplo de uso da função com dados fictícios
# Definindo alguns pontos fictícios
face = [
    Point(100, 100)
    Point(140, 100)
    Point(180, 100)
    # Pontos da boca (exemplo de pontos fictícios à esquerda e à direita)
    Point(385, 380), Point(387, 373)
]

# Índices de pontos da boca
p_boca_esq = [0, 1, 2, 3]  # Índices fictícios para a boca esquerda
p_boca_dir = [8, 9, 10, 11]  # Índices fictícios para a boca direita

# Chamar a função para calcular o MAR
distancia = calculo_ear(face, p_boca_esq, p_boca_dir)

print(f"A distância (MAR) entre a boca esquerda e direita é: {distancia}")
```
