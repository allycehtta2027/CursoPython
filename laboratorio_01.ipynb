# Águas de Março Analisando Palavras

```python
# exercicios - usando word_tokenize,
# tokenize apenas as palavras agora, na musica Águas de Março do Tom Jobim

palavras_agua_marco = word_tokenize(texto, language='portuguese')
print(palavras_agua_marco)
```

# Estatística(Probabilidade) - Ánalise Descritiva


# Frequência de distribuição

- Palvras com mais repetições

```python
from nltk.probability import FreqDist
```

```python
fdisk_aguas_marco = FreqDist(palavras_agua_marco)
```

```python
fdisk_aguas_marco
```

```python
fdisk_aguas_marco.most_common(2)
```

```python
fdisk_aguas_marco.plot(10)
```
